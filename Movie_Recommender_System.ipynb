we will first convert string to list
To convert from it, AST is used in python, under literal_eval of ast, it does the same thing.


```python
import ast
```

```python
#making a helper function to perform this action
def convert(data):
    List = []
    for i in ast.literal_eval(data):
        List.append(i['name']) 
    return List
```

```python
movie['genres'] = movie['genres'].apply(convert)
movie.head()
```

```python
# now after genres we will perform preprocessing on keywords
movie.iloc[1].keywords
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head()
```

```python
movie.iloc[1].cast
```

```python
# From Dataset in Cast coulmn we are taking only top 5 cast from all movie.
def change(data):
    List = []
    counter = 0
    for i in ast.literal_eval(data):
        if counter != 5:
            List.append(i["name"])
            counter+=1
        else:
          break
    return List 
```

```python
movie['cast'] = movie['cast'].apply(change)
movie.head()
```

```python
movie.iloc[1].crew
```

```python

# mujhe sirf  crew wale dataset wahi job role chahiye jaha director ka job specify kiya hai
def fetch_director(data):
    List = []
    for i in ast.literal_eval(data):
        if i['job'] == 'Director':
           List.append(i['name'])
           break
    return List
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
movie.head()
```

```python
movie['overview'][0]
```

```python
# all columns are in LIST except OVERVIEW.
#Overview is in the form of string so we will convert from String to LIST

movie['overview'] = movie['overview'].apply(lambda x:x.split())
movie.head()
```


# Now we have to remove space between the words 
If we keep these words separately then there will be problem like sam wargindom then machine will understand it as 2 tags problem tab will come when we are searching any movie and there is sam wargindom and sam mendes also here our machine will get confused which one is recommended Well, if we are looking for a sam wargindom movie and instead our recommendation system starts recommending sam Mendes' movie, then there will be a problem.
Example :- 'sam mendes' to 'sammendes'

# We will perform this action on the entire dataset.


```python
def collapse_space_word(text):
    List = []
    for i in text:
        List.append(i.replace(" ",""))
    return List
```

```python
movie['genres'] = movie['genres'].apply(collapse_space_word)
movie['keywords'] = movie['keywords'].apply(collapse_space_word)
movie['cast'] = movie['cast'].apply(collapse_space_word)
movie['crew'] = movie['crew'].apply(collapse_space_word)
movie.head()
```

We have to concatinate the 5 columns make one TAG name comumn.

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
movie.head()
```

```python
new_df = movie[['movie_id','title','tags']]
```

```python
new_df
```

# We have to remove (,) from Tags column

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

# Converting higher case to the lower case

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
         y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

Now we will do TEXT VECTORIZATION
And we have to calculate similarity between tags(2 tags ke under jo words hai unke beech ki similarity hume calculate karni hai)

for this we convert the text in to the vector
all movie are one vector
and in that we will take closest vector

we will use BagofW words
in bagofwords we will combine all the Tags
and fir usme hm word ka frequency calculate karenge kaunsa word kitne baar aaya hai.
jiska frequency sabse jyada hai us word ko extract karenge

we are going to use sckitlearn library

```python
from sklearn.feature_extraction.text import CountVectorizer
#countVectorizer(max_feature,stop_wards)
count_V = CountVectorizer(max_features = 5000 ,stop_words = 'english')
```

```python
vectors = count_V.fit_transform(new_df['tags']).toarray()
```

Har movie apne aap me ek vector hai to
har movie har dusre movie ke sath distance calculate karegi .
Distance jitna jyada hoga similarity utna kum hoga. Instead of finding EUCLEDIAN Distance we find Cosine Similarity ( we will find Angle between them jitna jyada angle utna disimilar hoga movie jitna angle kum hoga utna he jyada similar hoga. )
# In higher dimension Euclidean distance fail thats why we will use Cosine Similarity.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
#This is telling the similarity of the first movie with every movie.
similarity[0]
```

enumerate ki madad se movie apana position handle kar sakenge

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_df.iloc[i[0]].title)
```

This is similar movie to related our query

```python
recommend('Batman Begins')
```
